```python
import torch
```

```python
CPT_CONFIG_124M = {
    "vocab_size": 50257,
    "context_length": 1024,
    "emb_dim": 768,
    "n_heads": 12,
    "n_layers": 12,
    "drop_rate": 0.1,
    "qkv_bias": False,
}
```

```python
from ch4_utils import GPTModel, generate_text_simple
```

```python
model = GPTModel(CPT_CONFIG_124M)
```

```python
from tiktoken import get_encoding

tokenizer = get_encoding("gpt2")
```

```python
model.eval()
start_context = "Hello, I am"
encoded = tokenizer.encode(start_context)
print("encoded:", encoded)
encoded_tensor = torch.tensor(encoded).unsqueeze(0)
print(encoded_tensor.shape)
```

```python
out = generate_text_simple(
    model,
    encoded_tensor,
    max_new_tokens=6,
    context_size=CPT_CONFIG_124M["context_length"],
)
```

```python
decoded = tokenizer.decode(out[0].tolist())
```
